```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline 

path_to_data = 'data/'
output_path = 'data/clean/works_clean.csv'

w_df = pd.read_csv(path_to_data +'works.csv')
```

```python
w_df[:10]
```

## Drop:

#### Unnecessary columns

```python
w_df = w_df.drop(columns = ['ID', 'conductorName', 'soloists', 'orchestra', 'id'])
```

#### Rows with intermission

```python
w_df = w_df[w_df.interval != 'Intermission']
w_df = w_df[w_df.interval != 'Intermission-Short']
w_df = w_df.drop(columns = ['interval'])
```

#### Different movements of same work 

```python
w_df = w_df.drop(columns = ['movement'])
w_df = w_df.drop_duplicates()
```

```python
w_df[:10]
```

#### Write to file

```python
w_df.to_csv(output_path)
```
